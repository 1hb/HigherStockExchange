```python
# Следующая Функция забирает лист result с характеристиками приказов наших ботов. На выходе получается словарь с 3-мя
# различными датафреймами: market, change и limit, в которых содержится информация с соответствующими приказами
def queueing (result):
# Создаем List_of_Actions ботов с пустыми значениями 
    na_list = []
    i = 1
    while i <= len(USERS):
        na_list.append('NaN')
        i += 1
    na_list
    List_of_Actions = dict(zip(USERS, na_list))
    
# Заполняем List_Of_Actions действиями ботов после выполнения ими их функцих (результаты помещены в лист result)    
    keym = []
    for key in List_of_Actions:
        keym.append(key)
    i = 0
    while i <= (len(USERS)-1):
        List_of_Actions.update({keym[i] : result[i].get('action')})
        i += 1
        
# Создаем пустой словарь Temp_Data
    global Temp_Data
    Temp_Data = dict(zip(USERS, na_list))
    
# Заполняем Temp_Data позициями ботов после выполнения ими функций (цена + позиция)
    keym2 = []
    for key in Temp_Data:
        keym2.append(key)
    i = 0
    while i <= (len(USERS)-1):
        Temp_Data.update({keym2[i] : result[i].get('temp_data')})
        i += 1
        
# В List_of_Actions создаем в каждом датафрейме (value) столбец ID бота (его имя)
    for i in range(0, len(keym)):
        List_of_Actions[keym[i]]['ID'] = keym[i]
    List_of_Actions
    
# Объединяем все действия ботов в 1 датафрейм df_W
    df_w = pd.concat(List_of_Actions)
    
# Меняем индекс датафрейма df_w на числовой
    list1 = []
    for i in range(len(df_w)):
        list1.append(i)
    list1
    df_w.index = list1
    df_w
    
# Создаем словарь OrderQueue с 3 пустыми value
    OrderQueue = {'market' : 'NaN', 'change' : 'NaN', 'limit' : 'NaN'}

# Заполняем датафрейм market словаря df_w рыночными заявками роботов
    OrderQueue['market'] = df_w[(df_w.action == 0)&(df_w.buysell != 'Nan')&(df_w.volume > 0)&(df_w.volume != 'Nan')]

    list2 = []
    for i in range(len(OrderQueue['market'])):
        list2.append(i)
    OrderQueue['market'].index = list2

    if len(OrderQueue['market']) != 0:
        OrderQueue['market'].loc[:, 'no'] = 'NaN'
        OrderQueue['market'].loc[0, 'no'] = 1
        i = 1
        while i <= len(OrderQueue['market'].loc[:, 'ID']) - 1:
            if OrderQueue['market'].loc[i, 'ID'] == OrderQueue['market'].loc[i-1, 'ID']:
                OrderQueue['market'].loc[i, 'no'] = OrderQueue['market'].loc[i-1, 'no'] + 1
            else:
                OrderQueue['market'].loc[i, 'no'] = 1
            i = i + 1
    
        OrderQueue['market'] = OrderQueue['market'].sort_values('no').sort_values('ID')
        OrderQueue['market'] = OrderQueue['market'][['no','ID', 'buysell', 'volume']]

# Заполняем датафрейм limit словаря df_w лимитными заявками роботов
    OrderQueue['limit'] = df_w[(df_w.action == 1)&(df_w.buysell != 'Nan')&(df_w.volume != 'Nan')&(df_w.price != 'Nan')&(df_w.volume > 0)&(df_w.price > 0)]

    list4 = []
    for i in range(len(OrderQueue['limit'])):
        list4.append(i)
    OrderQueue['limit'].index = list4

    if len(OrderQueue['limit']) != 0:
        OrderQueue['limit'].loc[:, 'no'] = 'NaN'
        OrderQueue['limit'].loc[0, 'no'] = 1
        i = 1
        while i <= len(OrderQueue['limit'].loc[:, 'ID']) - 1:
            if OrderQueue['limit'].loc[i, 'ID'] == OrderQueue['limit'].loc[i-1, 'ID']:
                OrderQueue['limit'].loc[i, 'no'] = OrderQueue['limit'].loc[i-1, 'no'] + 1
            else:
                OrderQueue['limit'].loc[i, 'no'] = 1
            i = i + 1
    
        OrderQueue['limit'] = OrderQueue['limit'].sort_values('no').sort_values('ID')
        OrderQueue['limit'] = OrderQueue['limit'][['no','ID', 'buysell','price', 'volume']]
        
# Заполняем датафрейм change словаря df_w отзывными заявками роботов
    OrderQueue['change'] = df_w[(df_w.action == 2)&(df_w.buysell != 'Nan')&(df_w.orderno != 'Nan')&(df_w.volume != 'Nan')&(df_w.price != 'Nan')&(df_w.volume > 0)&(df_w.price > 0)]

    list3 = []
    for i in range(len(OrderQueue['change'])):
        list3.append(i)
    OrderQueue['change'].index = list3

    if len(OrderQueue['change']) != 0:
        OrderQueue['change'].loc[:, 'no'] = 'NaN'
        OrderQueue['change'].loc[0, 'no'] = 1
        i = 1
        while i <= len(OrderQueue['change'].loc[:, 'ID']) - 1:
            if OrderQueue['change'].loc[i, 'ID'] == OrderQueue['change'].loc[i-1, 'ID']:
                OrderQueue['change'].loc[i, 'no'] = OrderQueue['change'].loc[i-1, 'no'] + 1
            else:
                OrderQueue['change'].loc[i, 'no'] = 1
            i = i + 1
    
        OrderQueue['change'] = OrderQueue['change'].sort_values('no').sort_values('ID')
        OrderQueue['change'] = OrderQueue['change'][['no','ID', 'orderno', 'buysell','price', 'volume']]
        
    return(OrderQueue)
```
