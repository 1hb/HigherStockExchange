```python
import numpy as np
def bot_trading(trade_sec, last_book = 482114, last_trade = 122719):
    df_r <- Big_OrderBook[Big_OrderBook['TRADE_SEC'] == trade_sec]
    if len(df_r) == 0: return()   
 
    for i in range(1,len(df_r)):
        bot_order = df_r.iloc[i]
            # следующая строка нужна, если бы я использовал iloc, но т.к. я просто делаю append, она по сути не нужна
            # i_book = len(OrderBook[OrderBook['no'] != 0]) + 1
        orderno = OrderBook['orderno'].values.max() + 1    
        if bot_order['action'] == 1:             
            OrderBook.append(pd.DataFrame({'no': [np.append(OrderBook['no'].values, np.array(last_book)).max() + 1],
                                'time': [bot_order['time']],
                                'trade_sec': [bot_order['trade_sec']],
                                'ID': ['bot'],
                                'action': ['set'],
                                'buysell': [bot_order['buysell']],
                                'orderno': [orderno],
                                'bot_orderno': [bot_order['orderno']],
                                'price': [bot_order['price']],
                                'volume': [bot_order['volume']]
                               }))                                             
    
            
            if bot_order['buysell'] == 'B':
                DOM['demand'] = DOM['demand'].append(pd.DataFrame({'orderno': [orderno],
                                'bot_orderno': [bot_order['bot_orderno']],
                                'ID': [bot_order['ID']],
                                'price': [bot_order['price']],
                                'volume': [bot_order['volume']]
                               }))
                DOM['demand'] = DOM['demand'][DOM['demand']['volume'] != 0].sort_values(by = ['price', 'orderno'],
                                                                                       ascending = [False, True])
            if bot_order['buysell'] == 'S':
                DOM['supply'] = DOM['supply'].append(pd.DataFrame({'orderno': [orderno],
                                'bot_orderno': [bot_order['bot_orderno']],
                                'ID': [bot_order['ID']],
                                'price': [bot_order['price']],
                                'volume': [bot_order['volume']]
                               }))
                DOM['supply'] = DOM['supply'][DOM['supply']['volume'] != 0].sort_values(by = ['price', 'orderno'])
            
            while DOM['demand'].loc[0, 'price'] >= DOM['supply'].loc[0, 'price'] and len(DOM['demand'] != 0) and len(DOM['supply'] != 0):
                
                # добавляем новую строку в трейдбук
                vol = np.array([DOM['demand'].loc[0, 'volume'], DOM['supply'].loc[0, 'volume']]).min()
                TradeBook.append(pd.DataFrame({'tradeno': [np.append(TradeBook['no'].values, np.array(last_trade)).max() + 1],
                            'time': [bot_order['time']],
                            'trade_sec': [bot_order['trade_sec']],
                            'buy_ID': [DOM['demand'].loc[0, 'ID']],
                            'sell_ID': [DOM['supply'].loc[0, 'ID']],
                            'buy_orderno': [DOM['demand'].loc[0, 'orderno']],
                            'sell_orderno': [DOM['supply'].loc[0, 'orderno']],
                            'price': [DOM['demand'].loc[1, 'price'] if DOM['demand'].loc[0, 'orderno'] < DOM['supply'].loc[0, 'orderno'] else DOM['supply'].loc[1, 'price']],
                            'volume': [vol]
                           }))
                
                DOM['demand'].loc[1, 'volume'] -= vol
                DOM['supply'].loc[1, 'volume'] -= vol
                
                DOM['demand'] = DOM['demand'][DOM['demand']['volume'] != 0].sort_values(by == ['price', 'orderno'],
                                                                                       ascending = [False, True])
                DOM['supply'] = DOM['supply'][DOM['supply']['volume'] != 0].sort_values(by == ['price', 'orderno'])
                
                OrderBook.append(pd.DataFrame({'no': [np.append(OrderBook['no'].values, np.array(last_book)).max() + 1,
                                                     np.append(OrderBook['no'].values, np.array(last_book)).max() + 2],
                                'time': 2 * [bot_order['time']],
                                'trade_sec': 2 * [bot_order['trade_sec']],
                                'ID': [DOM['demand'].loc[0, 'ID'], DOM['supply'].loc[0, 'ID']],
                                'action': 2 * ['trade'],
                                'buysell': ['B', 'S'],
                                'orderno': [DOM['demand'].loc[0, 'orderno'], DOM['supply'].loc[0, 'orderno']],
                                'bot_orderno': [0, 0],
                                'price': 2 * [DOM['demand'].loc[1, 'price'] if DOM['demand'].loc[0, 'orderno'] < DOM['supply'].loc[0, 'orderno'] else DOM['supply'].loc[1, 'price']],
                                'volume': 2 * [vol]
                               }))
        if bot_order['action'] == 1:
            # проверяем на наличие заявки в стакане
            
            
           if (~np.isnan(ifelse(OrderBook[0, "buysell"] == "B",
                        DOM['demand'].loc[DOM['demand'].loc[0, 'orderno'].fillna(0), 'orderno'] == bot_order['orderno'],
                        DOM['supply'].loc[DOM['supply'].loc[0, 'orderno'].fillna(0), 'orderno'] == bot_order['orderno']))):                            
            OrderBook.append(pd.DataFrame({'no': [np.append(OrderBook['no'].values, np.array(last_book)).max() + 1],
                                'time': [bot_order['time']],
                                'trade_sec': [bot_order['trade_sec']],
                                'ID': ['bot'],
                                'action': ['drop'],
                                'buysell': [bot_order['buysell']],
                                'orderno': [ifelse(OrderBook[0, "buysell"] == "B", 
                                                            DOM['demand'].loc[DOM['demand'].loc[0, 'orderno'].fillna(0), 'orderno'] == bot_order['orderno'],
                                                            DOM['supply'].loc[DOM['supply'].loc[0, 'orderno'].fillna(0), 'orderno'] == bot_order['orderno'])],
                                'bot_orderno': [bot_order['orderno']],
                                'price': [bot_order['price']],
                                'volume': [bot_order['volume']]
                               }))                          
            if bot_order['BUYSELL'] == 'B':
                DOM['demand'] == DOM['demand']['orderno' != int(OrderBook[0, "orderno"])]
                         
                
            if bot_order['BUYSELL'] == 'S':
                DOM['supply'] == DOM['supply']['orderno' != int(OrderBook[0, "orderno"])]
                
                
    return OrderBook, TradeBook, DOM
```
