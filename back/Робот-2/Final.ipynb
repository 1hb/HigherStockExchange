```python
# Установка входных данных 
# Изначальный капитал
capit=1000000
#  При покупке данного типа акций нельзя тратить более 30% от своего изначального капитала
share_b=0.3
# Изначальный объем актива (нулевой)
vol_s=0
# Расчитываем волатильность по часовым ценам закрытия за рассматриваемый период
sd=np.std(df['<CLOSE>'])
# Устанавливаем целевую волатильность, выше которой используется стратегия STOCH-D+pe, ниже-SMA+pe
target_sd=30
# EPS за 2018 65.89 г. (последние доступные данные)
EPS = 65.89
df['<pe>'] = df['<CLOSE>']/EPS
pe_industry = 4.6 # Среднеотраслевое значение мультипликатора
```

```python
# Записываем условие выбора стратегии в зависимоти от волатильности за рассматриваемый период (в данном случае - год)
# Если волатильность больше целевого уровня используем стратегию STOCH-D+pe
for i in range(1, len(df['<CLOSE>'])):
    if sd>target_sd:
        df['<Return>'] = df['<CLOSE>'] / df['<CLOSE>'].shift()-1 
        df['<LL>'] = df['<LOW>'].rolling(window=15).min()
        df['<HH>'] = df['<HIGH>'].rolling(window=15).max()
        df['<%K>'] = 100*((df['<CLOSE>'] - df['<LL>']) / (df['<HH>'] - df['<LL>']))
        df['<%D>'] = df['<%K>'].rolling(window=7).mean()
        df['<Buy>'] = ((df['<%K>'] > df['<%D>']) & (df['<%K>'].shift(1) < df['<%D>'].shift(1))) & (df['<%D>'] <= 20) & (df['<pe>'] < pe_industry)
        df['<Sell>'] = ((df['<%K>'] < df['<%D>']) & (df['<%K>'].shift(1) > df['<%D>'].shift(1))) & (df['<%D>'] >= 80)
        df['<%K>'] = df['<%K>'].fillna(0)
        df['<%D>'] = df['<%D>'].fillna(0)
        df['<Sign>'] = 0
        df.loc[df['<Buy>'],'<Sign>'] = 1 
        df.loc[df['<Sell>'],'<Sign>'] = -1
        df['<st_return>'] = df['<Return>'] * df['<Sign>']
        df['<st_return>'] = df['<st_return>'].fillna(0)
# Если волатильность ниже целевого уровня используем стратегию SMA+pe
    elif sd<target_sd: 
        df['<SMA>']= df['<CLOSE>'].rolling(window=12).mean()
        df['<RETURN_SMA>'] = df['<OPEN>']/ (df['<CLOSE>'].shift(-1))-1
        df['<RETURN_1>'] = df['<OPEN>'] /(df['<CLOSE>'].shift(-2))-1
        df['<RETURN_2>']=df['<CLOSE>']/df['<CLOSE>'].shift(-1)-1
        df['<Buy>'] = ((df['<SMA>'] > df['<CLOSE>']) & (df['<SMA>'].shift(-1) <= df['<CLOSE>'].shift(-1)) & (df['<pe>']<pe_industry))
        df['<Sell>'] = ((df['<SMA>'] < df['<CLOSE>']) & (df['<SMA>'].shift(-1) >= df['<CLOSE>'].shift(-1)))
        df.loc[df['<Buy>'],'<Sign>'] = 1 
        df.loc[df['<Sell>'],'<Sign>'] = -1
        df['<st_return>'] = df['<RETURN_SMA>'] * df['<Sign>']
        df['<st_return>'] = df['<RETURN_SMA>'].fillna(0)

df.head(15)
```
